# Color Spaces

### Import the necessary packages

```python
import matplotlib.pyplot as plt
import numpy as np
import argparse
import cv2

%matplotlib inline
```

### Load the image and show it

```python
image = cv2.imread("images/image2.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) #Converting to RGB (from BGR)
plt.imshow(image)
```

### Convert the image to grayscale

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

### Convert the image to the HSV (Hue, Saturation, Value) color spaces

```python
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
plt.imshow(hsv)
```

### Convert the image to the L x a x b color spaces

```python
lab = cv2.cvtColor(image, cv2.COLOR_RGB2LAB)
plt.imshow(lab)
```
