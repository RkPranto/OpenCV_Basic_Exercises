# Image Basics

### Import the necessary packages

```python
from __future__ import print_function
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

### Load the image and show it

```python
image = cv2.imread("images/image1.jpg")
plt.imshow(image)
```

### Images are just NumPy arrays. The top-left pixel can be found at (0, 0)

```python
(b, g, r) = image[0, 0]
print("Pixel at (0, 0) - Red: {}, Green: {}, Blue: {}".format(r, g, b))
```

### Now, let's change the value of the pixel at (0, 0) and make it red

```python
image[0, 0] = (0, 0, 255)
(b, g, r) = image[0, 0]
print("Pixel at (0, 0) - Red: {}, Green: {}, Blue: {}".format(r, g, b))
```

### Since we are using NumPy arrays, we can apply slicing and grab large chunks of the image. Let's grab the top-left corner

```python
corner = image[0:100, 0:100]
plt.imshow(corner)
```

### Let's make the top-left corner of the image green

```python
image[0:100, 0:100] = (0, 255, 0)
```

### Show our updated image

```python
plt.imshow(image)
```
